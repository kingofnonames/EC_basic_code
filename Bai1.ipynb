### Tạo quần thể theo phân phối uniform

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def create_population(pop_size=20, dim=10, lower=-5, upper=5):
    population = np.random.uniform(lower, upper, (pop_size, dim))
    return population

def fitness_function(individual):
    return np.sum(individual ** 2)
```

```python
pop = create_population(pop_size=5, dim=4, lower=-5, upper=5)
print("Quần thể khởi tạo:")
print(pop)
```

```python
fitness_values = [fitness_function(individual) for individual in pop]
print("\nFitness của từng cá thể:")
for i, fit in enumerate(fitness_values):
    print(f"Cá thể {i}: fitness = {fit:.4f}")
```

### Vẽ demo hàm Sphere, Rastrigin, Ackley 2D

```python
def sphere(x, y):
    return x**2 + y**2

def rastrigin(x, y, A=10):
    return 2*A + (x**2 - A*np.cos(2*np.pi*x)) + (y**2 - A*np.cos(2*np.pi*y))

def ackley(x, y, a=20, b=0.2, c=2*np.pi):
    part1 = -a * np.exp(-b * np.sqrt(0.5*(x**2 + y**2)))
    part2 = -np.exp(0.5*(np.cos(c*x) + np.cos(c*y)))
    return part1 + part2 + a + np.exp(1)

def plot_function(func, name, x_range=(-5, 5), y_range=(-5, 5), step=0.1):
    x = np.arange(x_range[0], x_range[1], step)
    y = np.arange(y_range[0], y_range[1], step)
    X, Y = np.meshgrid(x, y)
    Z = func(X, Y)
    plt.figure(figsize=(6, 5))
    cp = plt.contour(X, Y, Z, levels=50, cmap='viridis')
    plt.colorbar(cp)
    plt.title(name)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.show()


plot_function(sphere, "Sphere Function")
plot_function(rastrigin, "Rastrigin Function")
plot_function(ackley, "Ackley Function")
```
